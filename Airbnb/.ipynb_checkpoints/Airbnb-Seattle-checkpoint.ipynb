# Airbnb Activity Data Analysis

### 1. Business Understanding

Airbnb has become a top choice for travelers who look for short-term accomodations because it provides not only cheaper service than conventional hotels, but also personalized experience that benefits from its peer-to-peer business model. On the other hand, Airbnb provides an opportunity for the hosts to earn profit.

An airbnb host may be curious about how the market looks like and how to improve the business to earn more money. Based on the Airbnb activity data, this project focused on three questions that an airbnb host in Seattle may want to know before making strategic decisions:

1. What does the market look like across the neighborhoods in Seattle?

2. What are the best months for the hosts?

3. What features of a property/listing attract more bookings?

### 2. Data Understanding

The Seattle Airbnb Open Data describe the listing activity of homestays in Seattle, WA. There are three dataset included:

1. Calendar dataset includes the listing id, the price and the availability for that day.
2. Listings dataset includes the full descriptions of the listings and the average review score.
3. Reviews dataset includes the unique id for each reviewer and the detailed comments.

### 3. Data Preparation

Based on the overview of the datasets, this section performs some data wrangling to prepare the data we need for further analysis.

#### Calculate the occupancy rate

The occupancy rate is the number of days during a specific time period in which a property is rented out by guests. The availability data are provided in the calendar dataset. The listings with 0 occupancy rate are considered "not active" and are filtered out in this project.

```python
calendar_total = calendar.groupby(['listing_id', 'available'])['date'].count().reset_index(name = 'count')
calendar_t_1 = calendar_total[calendar_total['available'] == 't']
calendar_f = calendar_total[calendar_total['available'] == 'f']
calendar_t_2 = calendar_f[calendar_f['count'] == 365]
calendar_t_2['count'] = 0
calendar_t_2['available'] = 't'

availability = pd.concat([calendar_t_1, calendar_t_2])
availability['availability'] = availability['count'] / 365
availability.sort_values(by = ['availability'], ascending = False, inplace = True)
availability['Occupancy'] = 1 - availability['availability']
availability = availability[(availability['Occupancy'] > 0) & (availability['Occupancy'] <= 1)]
```

#### Convert price and fees from object type to float type

The price and the cleaning fee in the data are stored as object type with a dollar sign as prefix. We need to remove the dollar sign and convert the data to numerical data for further analysis.

```python
def get_price(val):
'''
Remove dollar sign and convert price to numeric numbers.

Args:
    string: the price with a dollar sign and separator
    
Returns: 
    float: the price that only keeps the numbers in the input argument
    
'''
#     print(val)
    if val != val:
        return val
    val = float(val.replace('$', '').replace(',', ''))
    return val
```

```python
calendar['price'] = calendar['price'].apply(lambda x: get_price(x))
```

```python
listings['price'] = listings['price'].apply(lambda x: get_price(x))
listings['cleaning_fee'] = listings['cleaning_fee'].apply(lambda x: get_price(x))
```

#### Convert rates from object type to float type

Similar to the price, the rate column is also in object type with percentage sign in it. We need to convert the column to float type.

```python
def percentage(val):
"""
Convert the percentage values to numerical values.

Args:
    string: a percentage value with % sign
    
Returns:
    float: fractional percentage value
"""
    if not pd.isnull(val):
        val = float(val.replace('%', ''))/100
    return val

listings['host_response_rate'] = listings['host_response_rate'].apply(lambda x: percentage(x))
listings['host_acceptance_rate'] = listings['host_acceptance_rate'].apply(lambda x: percentage(x))
```

#### Dealing the categorical variables and missing values

Encode the categorical variables

For the binary variables, encode True as 1, False as 0.

```python
def boolean_int(val):
"""
Encode binary variables

Args:
    boolean: True/False

Returns:
    int: 1 for True, 0 for False
"""
    if val == 't':
        return 1
    if val == 'f':
        return 0
    
listings_sub['host_is_superhost'] = listings_sub['host_is_superhost'].apply(lambda x: boolean_int(x))
listings_sub['host_identity_verified'] = listings_sub['host_identity_verified'].apply(lambda x: boolean_int(x))
listings_sub['host_has_profile_pic'] = listings_sub['host_has_profile_pic'].apply(lambda x: boolean_int(x))
listings_sub['is_location_exact'] = listings_sub['is_location_exact'].apply(lambda x: boolean_int(x))
```

For the ordinal categorical variables, encode the values to integers with specific order.

```python
# convert response time
response_time_dict = {'within an hour': 1, 'within a few hours': 2,
             'within a day': 3, 'a few days or more': 4}
listings_sub['host_response_time'] = listings_sub['host_response_time'].apply(lambda x: response_time_dict[x] if (pd.notnull(x)) else x)

# convert cancellation policy
cancellation_dict = {'flexible': 1, 'moderate': 2, 'strict': 3}
listings_sub['cancellation_policy'] = listings_sub['cancellation_policy'].apply(lambda x: cancellation_dict[x] if (pd.notnull(x)) else x)
```

For the numerical variables, the missing values are imputed using the mean. For the categorical variables, we will first encode them, then impute the missing values using the mode.

Check missing values

```python
listings_sub.isna().sum()
```

```python
category_list = ['bed_type', 'amenities', 'property_type', 'room_type', 'cancellation_policy',
                'host_identity_verified', 'host_has_profile_pic', 'is_location_exact', 'host_is_superhost']

numerical_list = ['host_response_rate', 'host_acceptance_rate', 'accommodates', 'bathrooms', 'bedrooms', 'beds', 'cleaning_fee', 
                 'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness',
                  'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 
                  'review_scores_value', 'number_of_reviews', 'price']

# fill nan values for numerical columns
listings_sub[numerical_list] = listings_sub[numerical_list].fillna(listings_sub[numerical_list].mean())

# fill nan values for categorical columns
cols = ['host_is_superhost', 'host_identity_verified', 'host_has_profile_pic',
        'host_response_time', 'cancellation_policy', 'property_type']
listings_sub[cols] = listings_sub[cols].fillna(listings_sub[cols].mode().iloc[0])
```

```python
# merge occupancy rate into the subset
listings_sub = pd.merge(listings_sub, availability, left_on = 'id', right_on = 'listing_id')
```

#### Question 3: What features of a property/listing attract more bookings?

First, the correlation matrix between occupancy rate and selected features is computed and visualized. The feature matrix is normalized to make sure the variables are in the same scale.

```python
# occupancy corr
var = np.array(listings_sub[['Occupancy', 'host_has_profile_pic', 'cancellation_policy',
                        'host_response_rate', 'bedrooms','review_scores_rating', 'cleaning_fee']]).reshape(-1, 7)
var = (var-var.min(axis=0))/(var.max(axis=0)-var.min(axis=0))
var_df = pd.DataFrame(var, columns = ['Occupancy', 'host_has_profile_pic', 'cancellation_policy',
                        'host_response_rate', 'bedrooms','review_scores_rating', 'cleaning_fee'])
corr = var_df.corr()

sns.set(style = 'white')
mask = np.zeros_like(corr, dtype = np.bool)
mask[np.triu_indices_from(mask)] = True
cmap = sns.diverging_palette(220, 10, as_cmap = True)
sns.heatmap(corr, cmap = cmap, mask = mask, vmax = 0.3, center = 0, 
            linewidths = 0.5, square = True, cbar_kws = {'shrink': 0.5})
```

##### Quantitative analysis - modeling

A linear regression is conducted to provide a quantitative relationship between the occupancy rate and the selected numerical features. 

```python
Y = np.array(listings_sub['Occupancy']).reshape(-1, 1)
X = np.array(listings_sub[['host_identity_verified', 'host_has_profile_pic', 'cancellation_policy',
                        'host_response_rate', 'host_response_time',
                        'accommodates', 'bedrooms', 'cleaning_fee',
                        'review_scores_rating', 'host_is_superhost']]).reshape(-1, 10)
X = (X-X.min(axis=0))/(X.max(axis=0)-X.min(axis=0))
X_const = sm.add_constant(X) # add a column of ones to the independent variable, i.e., add the intercept term in the model
model_stats = sm.OLS(Y, X_const).fit()

# the output of the fitted model
print(model_stats.summary())
```

With a 5% significance level, the result shows that most variables are not statistically significant in the linear model. However, the cancellation policy, number of bedrooms, and the cleaning fee have estimated coefficient that we can trust is not zero. 

Both the correlation visualization and the linear model show that strict cancellation policy and high cleaning fee will negatively affect the occupancy rate.

Another feature that may correlated with the occupancy rate is the room type and the property type, both of which are nominal variables. The market share and the occupancy rate of different property types and room types.

Again, median occupancy rate is chosen as the representative occupancy rate for its group.

```python
room_count = listings_sub.groupby(['room_type'])['id'].count().reset_index(name = 'count')
room_count['perc'] = room_count['count']/len(listings_sub['id'].unique())
room_type = listings_sub.groupby(['room_type'])['Occupancy'].median().reset_index(name = 'Median Occupancy')
room_type = pd.merge(room_type, room_count, on = 'room_type')
```

The majority types in the market are apartment and house. But we will also include two less common property types, townhouse and condominium, in the analysis.

```python
property_count = listings_sub.groupby(['property_type'])['id'].count().reset_index(name = 'count')
property_count['perc'] = property_count['count']/len(listings_sub['id'].unique())
property_count.sort_values(by = ['perc'], ascending = False)[:10]
```

```python
property_count = property_count[(property_count['property_type'] == 'Apartment') |
                             (property_count['property_type'] == 'House') |
                             (property_count['property_type'] == 'Townhouse') |
                             (property_count['property_type'] == 'Condominium')]

property_type = listings_sub.groupby(['property_type'])['Occupancy'].median().reset_index(name = 'Median Occupancy')
property_type = property_type[(property_type['property_type'] == 'Apartment') |
                             (property_type['property_type'] == 'House') |
                             (property_type['property_type'] == 'Townhouse') |
                             (property_type['property_type'] == 'Condominium')]
property_type.sort_values(by = ['Median Occupancy'], inplace = True)
property_type = pd.merge(property_type, property_count, on = 'property_type')
```

```python
fig = plt.figure(figsize=(20, 7))
axs = []

# plot occupancy rate & market share for different property types
axs.append(fig.add_subplot(121))
axs[0].bar(property_type['property_type'], 100*property_type['Median Occupancy'],
       color = 'bisque', width = 0.3, label = 'Occupancy Rate')
axs[0].set_ylabel('Occupancy Rate', fontsize = 20)
axs[0].set_xlabel('Property Type', fontsize = 20)
plt.xticks(fontsize = 18)
plt.yticks(fontsize = 18)
axs[0].yaxis.set_major_formatter(mtick.PercentFormatter())

ax2 = axs[0].twinx()  # share the x-axis, new y-axis
color2 = 'burlywood'
ax2.set_ylabel('Market Share', fontsize=20)
ax2.plot(property_type['property_type'], 100*property_type['perc'], color = color2, 
         marker = 'D', ms = 8, label = 'Market Share')
ax2.tick_params(axis='y')
ax2.yaxis.set_major_formatter(mtick.PercentFormatter())
plt.xticks(fontsize = 18)
plt.yticks(fontsize = 18)

# plot occupancy rate & market share for different room types
axs.append(fig.add_subplot(122))
axs[1].bar(room_type['room_type'], 100*room_type['Median Occupancy'],
       color = 'bisque', width = 0.3, label = 'Occupancy Rate')
plt.xticks(fontsize = 18)
plt.yticks(fontsize = 18)
axs[1].set_ylabel('Occupancy Rate', fontsize = 20)
axs[1].set_xlabel('Room Type', fontsize = 20)
axs[1].yaxis.set_major_formatter(mtick.PercentFormatter())

ax3 = axs[1].twinx()  # share the x-axis, new y-axis
color3 = 'burlywood'
ax3.set_ylabel('Market Share', fontsize = 20)
ax3.plot(room_count['room_type'], 100*room_count['perc'], color = color3, 
         marker = 'D', ms = 8, label = 'Market Share')
ax3.tick_params(axis = 'y')
ax3.yaxis.set_major_formatter(mtick.PercentFormatter())
plt.xticks(fontsize = 18)
plt.yticks(fontsize = 18)
axs[1].legend(loc = 'center left', bbox_to_anchor = (1.2, 0.57), fontsize = 16)
ax3.legend(loc = 'center left', bbox_to_anchor = (1.2, 0.5), fontsize = 16)

plt.subplots_adjust(wspace = 0.46)
```

It turns out that customers prefer having the entire home compared to having only one private room or the shared room. In terms of the property type, even though apartment and house occupy the majority of the market, there is a chance that the rare types attract customers more and get much higher occupancy rate.

Following code extract the words from the listing titles and count the most frequently used words. We can compare the most frequent words used by super hosts and other hosts, listings with high occupancy rate versus low.

```python
def get_str(name):
"""Get the list of words in a string"""
    chars = re.escape(string.punctuation)
    return re.sub(r'['+chars+']', ' ', name)
```

```python
def get_name_freq(name):
"""Get the words occur in the string list"""
    words = ''
    synonym = dict({'Apt': 'Apartment', 'apt': 'Apartment', 'bdrm': 'Bedroom'})
    for i in range(len(name)):
        name_i = name.iloc[i]
        for j in range(len(name_i)):
            if name_i[j] in synonym.keys():
                name_i[j] = synonym[name_i[j]]
        words += ' '.join(name_i) + ' '
    return words
```
